Алгоритм:

Определение спектра из текстового файла
Создание функции для определения диапазона С=С и нахождения его максимума и минимума
Создание функции для определения диапазона С-N и нахождения его максимума и минимума
Создание функции для определения диапазона N-O и нахождения его максимума и минимума
Вывод диапазона характерных полос для каждой функциональной группы с координатами максимума и минимума
Построение графиков спектров, выделение характерных полос и отметка максимумов и минимумов

```python
import matplotlib as mp
import numpy as np
arr_x = [] #создание массива для значений х
arr_y = [] #создание массива для значений у
with open('coords.txt', 'r') as file: #открытие и чтение файла спектра
  for line in file.read().replace("\t", ", ").splitlines(): #цикл для соединения координат
    x, y = map(float, line[0:-1].split(", ")) #определение х и у 
    arr_x.append(x) #добавление значений х в массив arr_x
    arr_y.append(y) #добавление значений у в массив arr_у
#Диапазоны характерных полос следующие:
# 1) С=С от 1635 до 1690;
# 2) C-N от 1000 до 1400;
# 3) N-O от 1500 до 1560.
items_xCC = []
index_xCC = []
items_yCC = []
#функция для определения характерной полосы для С=С
def diapazon_cc():
  for index, item in enumerate(arr_x):
    if 1635 <= item <= 1690:
      items_xCC.append(item)
      index_xCC.append(index)
  for point_cc in index_xCC:
    items_yCC.append(arr_y[point_cc])
  return items_xCC, items_yCC
x_CC, y_CC = diapazon_cc()
items_xCN = []
index_xCN = []
items_yCN = []
#функция для определения характерной полосы для С-N
def diapazon_cn():
  for index, item in enumerate(arr_x):
    if 1000 <= item <= 1400:
      items_xCN.append(item)
      index_xCN.append(index)
  for point_cn in index_xCN:
    items_yCN.append(arr_y[point_cn])
  return items_xCN, items_yCN
x_CN, y_CN = diapazon_cn()
items_xNO = []
index_xNO = []
items_yNO = []
#функция для определения характерной полосы для N-O
def diapazon_no():
  for index, item in enumerate(arr_x):
    if 1500 <= item <= 1560:
      items_xNO.append(item)
      index_xNO.append(index)
  for point in index_xNO:
    items_yNO.append(arr_y[point])
  return items_xNO, items_yNO
x_NO, y_NO = diapazon_no()
#вывод диапазона характерных полос для каждой функциональной группы с координатами максимума
print('Range C=C from 1635 to 1690, maximum =', (x_CC[y_CC.index(max(y_CC))], max(y_CC)), ', minimum =', (x_CC[y_CC.index(min(y_CC))], min(y_CC)))
print('Range C-N from 1000 to 1400, maximum =', (x_CN[y_CN.index(max(y_CN))], max(y_CN)), ', minimum =', (x_CN[y_CN.index(min(y_CN))], min(y_CN)))
print('Range N-O from 1500 to 1560, maximum =',(x_NO[y_NO.index(max(y_NO))], max(y_NO)), ', minimum =', (x_NO[y_NO.index(min(y_NO))], min(y_NO)))
#отрисовка графика спектра
fig = mp.pyplot.figure(figsize=(100, 8))
mp.pyplot.scatter(x_CC[y_CC.index(max(y_CC))], max(y_CC), marker='o', c= 'r' )
mp.pyplot.scatter(x_CN[y_CN.index(max(y_CN))], max(y_CN), marker='o', c= 'purple' )
mp.pyplot.scatter(x_NO[y_NO.index(max(y_NO))], max(y_NO), marker='o', c= 'r' )
mp.pyplot.scatter(x_CC[y_CC.index(min(y_CC))], min(y_CC), marker='o', c= 'r' )
mp.pyplot.scatter(x_CN[y_CN.index(min(y_CN))], min(y_CN), marker='o', c= 'purple' )
mp.pyplot.scatter(x_NO[y_NO.index(min(y_NO))], min(y_NO), marker='o', c= 'r' )
ax = fig.add_subplot()
ax1 = fig.add_subplot()
ax2 = fig.add_subplot()
ax3 = fig.add_subplot()
ax.plot(arr_x, arr_y , label = 'Spectrum')
ax1.plot(x_NO, y_NO, c = 'black' , label = 'N-O')
ax2.plot(x_CC, y_CC, c= 'g', label ='C=C')
ax3.plot(x_CN, y_CN, c= 'r', label ='C-N')
ax.legend()
ax1.legend()
ax2.legend()
ax3.legend()
mp.pyplot.show()
```
